# Home Credit Default Risk

## Data Overview

```python
# Import Libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# connect to postgres
url = "postgresql://postgres:password@localhost:5433/final_project"
engine = create_engine(url)
```

```python
# extract train data
application_train = pd.read_sql('select * from home_credit_default_risk_application_train', con=engine)
application_train.head()
```

```python
# extract test data
application_test = pd.read_sql('select * from home_credit_default_risk_application_test', con=engine)
application_test.head()
```

```python
# drop columns if null value is more than 60%

drop_columns = []
for col in application_train.columns:
    null_pct = application_train[col].isna().sum() / len(application_train)
    if null_pct > 0.06:
        drop_columns.append(col)
```

```python
application_train_clean = application_train.drop(drop_columns, axis=1)
application_train_clean.head()
```

```python
application_test_clean = application_test.drop(drop_columns, axis=1)
application_test_clean.head()
```

```python
# Import clean data to database
application_train_clean.to_sql('home_credit_default_risk_application_train_clean', index=False, con=engine, if_exists='replace')
application_test_clean.to_sql('home_credit_default_risk_application_test_clean', index=False, con=engine, if_exists='replace')
```

# Preprocessing

```python
# Splitting data
X_train = application_train_clean.drop(columns=['TARGET'],axis=1)
y_train = application_train_clean['TARGET']
X_test = application_test_clean
```

```python
# Numerical Columns
num=X_train.select_dtypes(exclude='object').columns

# Categorical Columns
cat=X_train.select_dtypes(include='object').columns
```

# Logistic Regression Modelling

```python
# import libraries
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

from sklearn.metrics import precision_recall_fscore_support
```

```python
# Transformer and Modelling Steps
cat_transformer = Pipeline([
    ("c_i", SimpleImputer(strategy="most_frequent")),
    ("e", OneHotEncoder(handle_unknown = "ignore"))
])

num_transformer = Pipeline([
    ("n_i", SimpleImputer(strategy="median"))
])

transformers = [
    ("n_t", num_transformer, num),
    ("c_t", cat_transformer, cat)
]

model = Pipeline([
    ("pre", ColumnTransformer(transformers=transformers)),
    ("model", LogisticRegression(max_iter=1000))
])
```

```python
# Fitting model
model.fit(X_train, y_train)
```

```python
# Predict
y_test = model.predict(X_test)
```

```python
# Prediction Probability
y_prob = model.predict_proba(X_test)
```

# Upload ML result to postgres

```python
# combine the result of machine learning

# get id
ml_result = X_test[['SK_ID_CURR']]

# add ml result
ml_result['prediction_target'] = y_test.tolist()
ml_result['probability'] = y_prob.tolist()

ml_result
```

```python
# upload ml result to postgres
ml_result.to_sql('home_credit_default_risk_application_ml_result', index=False, con=engine, if_exists='replace')
```
